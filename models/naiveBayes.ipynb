```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import GaussianNB, MultinomialNB

from sklearn.metrics import accuracy_score
```

```python
train =  pd.read_csv('../train.csv')
test =  pd.read_csv('../test.csv')

print("Train data shape:", train.shape)
print("Test data shape:", test.shape)

train.drop(columns=['Unnamed: 0','id'],inplace = True)
test.drop(columns=['Unnamed: 0','id'],inplace = True)
Y = train['satisfaction']
X = train.drop('satisfaction', axis=1)
```

```python
df = pd.DataFrame(X)

# Apply LabelEncoder to each column
X = LabelEncoder()
X = df.apply(X.fit_transform)

# Print the transformed DataFrame
X
```

```python
df = pd.DataFrame(Y)

# Apply LabelEncoder to each column
Y = LabelEncoder()
Y = df.apply(Y.fit_transform)

# Print the transformed DataFrame
Y
```

```python

df = pd.DataFrame(test)

# Apply LabelEncoder to each column
enc = LabelEncoder()
test = df.apply(enc.fit_transform)

# Print the transformed DataFrame
test
y_test = test['satisfaction']
test = test.drop('satisfaction', axis=1)
```

```python
# Create a Naive Bayes classifier (Gaussian Naive Bayes)
nb = GaussianNB()

# Train the model using the training data
nb.fit(X, Y)

# Make predictions on the test data
y_pred = nb.predict(test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Create a Naive Bayes classifier (Gaussian Naive Bayes)
nb2 = MultinomialNB()

# Train the model using the training data
nb2.fit(X, Y)

# Make predictions on the test data
y_pred = nb2.predict(test)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
